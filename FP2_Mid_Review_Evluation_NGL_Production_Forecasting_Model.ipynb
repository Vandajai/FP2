```python
# Import the necessary Libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Create a new dataframe for the creation of the new variables
ngl_df_new = ngl_df

# Create the trend variable
ngl_df_new['Trend'] = np.arange(1, len(ngl_df_new)+1)

# Create the seasonality variable (from 1 to 12)
ngl_df_new['Month'] = ngl_df_new.index.month

# Display the 'ngl_df_new' dataset
ngl_df_new.head(24)
```

```python
# Split the data into training and validation sets
train = ngl_df_new[:-12]
valid = ngl_df_new[-12:]

# Display the observations of the train dataset
train.head()
```

```python
# Display the observations of the validation dataset
valid.head()
```

# Step 1: Fit Model A - Linear Trend with Additive Seasonality

For this model, we consider a linear trend and additive seasonality: <br>

Production_Value = 0 + 1Trend + month +  <br>

0 = Intercept <br>
1 = Coefficient of the linear trend <br>
month = Monthly seasonality <br>
 = error <br>

```python
# Prepare features for Model A
X_train_A = pd.get_dummies(train['Month'], prefix='Month').drop('Month_1', axis=1)
X_train_A['Trend'] = train['Trend']

# Fit the model
model_A = LinearRegression()
model_A.fit(X_train_A, train['Value'])

# Coefficients
print("Model A Coefficients:\n")
print("Intercept:", model_A.intercept_)
print("Trend:", model_A.coef_[-1])
print("Seasonality:", model_A.coef_[:-1])
```

# Step 2: Fit Model B - Exponential Trend with Multiplicative Seasonality

```python
# Prepare features for Model B
X_train_B = pd.get_dummies(train['Month'], prefix='Month').drop('Month_1', axis=1)
X_train_B['Trend'] = train['Trend']

# Fit the model on the log-transformed target variable
model_B = LinearRegression()
model_B.fit(X_train_B, np.log(train['Value']))

# Coefficients
print("Model B Coefficients:\n")
print("Intercept:", model_B.intercept_)
print("Trend:", model_B.coef_[-1])
print("Seasonality:", model_B.coef_[:-1])
```

```python
# Prepare features for Model A
X_train_A = pd.get_dummies(train['Month'], prefix='Month').drop('Month_1', axis=1)
X_train_A['Trend'] = train['Trend']

# Fit the model
model_A = LinearRegression()
model_A.fit(X_train_A, train['Value'])


# Prepare validation set features
X_valid_A = pd.get_dummies(valid['Month'], prefix='Month').drop('Month_1', axis=1)
X_valid_A['Trend'] = valid['Trend']

X_valid_B = pd.get_dummies(valid['Month'], prefix='Month').drop('Month_1', axis=1)
X_valid_B['Trend'] = valid['Trend']

# Model A Predictions
valid['Predict_A'] = model_A.predict(X_valid_A)

# Model B Predictions (and reverse the log transform)
valid['Predict_B'] = np.exp(model_B.predict(X_valid_B))

# RMSE
rmse_A = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_A']))
rmse_B = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_B']))

print(f"Validation RMSE for Model A: {rmse_A}")
print(f"Validation RMSE for Model B: {rmse_B}")
```

```python
# Compare RMSE of Model A and Model B on validation data
print(f"Validation RMSE for Model A: {rmse_A}")
print(f"Validation RMSE for Model B: {rmse_B}")

if rmse_A < rmse_B:
    print("Model A is better.")
else:
    print("Model B is better.")
```

```python
# Log transform the Sales for Model B
ngl_df_new['Log_Value'] = np.log(ngl_df_new['Value'])

# Extract features and target variable for training
X = ngl_df_new[['Month', 'Trend']]  # Based on the original dataframe
y = ngl_df_new['Log_Value']

# Fit the model to the training set
model_B_all = LinearRegression()
model_B_all.fit(X, y)

# Prepare the input for September 2023 prediction
last_trend_value = ngl_df_new['Trend'].iloc[-1]  # The last trend value from the data
X_sep_2023 = np.array([1, last_trend_value + 1]).reshape(1, -1)

# Make production value prediction for September 2023
log_value_sep_2023 = model_B_all.predict(X_sep_2023)

# Convert the log production value back to regular value scale
value_sep_2023 = np.exp(log_value_sep_2023)

print(f"The forecasted NGL production for September 2023 is approximately ${value_sep_2023[0]:.2f} kb/d")
```

```python
# Obtain predicted values for the training set
y_pred = model_B_all.predict(X)

# Calculate residuals of Model B
residuals = y - y_pred
```

```python
# Import the necessary libraries for acf and pacf plots
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF
plt.figure(figsize=(12, 4))
plot_acf(residuals, lags=20)
plt.title('ACF of Residuals')
plt.show()

# Plot PACF
plt.figure(figsize=(12, 4))
plot_pacf(residuals, lags=20)
plt.title('PACF of Residuals')
plt.show()
```

# Evaluate ARIMA (2,0,0) Model

```python
import statsmodels.api as sm

# Assuming train_residuals is a pandas Series containing residuals from the training set of Model B

# Fit AR(2) model to the residuals
model_AR2 = sm.tsa.ARIMA(residuals, order=(2, 0, 0)).fit()
print(model_AR2.summary())

# Produce ACF and PACF plots for the new residuals
new_residuals = model_AR2.resid
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# ACF
sm.graphics.tsa.plot_acf(new_residuals, lags=20, ax=ax[0])
ax[0].set_title('ACF of AR(2) Residuals')

# PACF
sm.graphics.tsa.plot_pacf(new_residuals, lags=20, ax=ax[1])
ax[1].set_title('PACF of AR(2) Residuals')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Helper function to compute MAPE
def compute_mape(y_true, y_pred):
    # Avoid division by zero
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    non_zero_mask = y_true != 0
    y_true = y_true[non_zero_mask]
    y_pred = y_pred[non_zero_mask]
    
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


# Compute RMSE for each combined forecast of Model A and Model B with AR(p_best) and the other ARIMA and SARIMA models
rmse_combined_Model_A = np.sqrt(mean_squared_error(valid['Value'], valid['Combined_Predict_Model_A']))
rmse_combined_Model_B = np.sqrt(mean_squared_error(valid['Value'], valid['Combined_Predict_Model_B']))
rmse_Model_A = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_A']))
rmse_Model_B = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_B']))
rmse_Model_1 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_1']))
rmse_Model_2 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_2']))
rmse_Model_3 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_3']))
rmse_Model_4 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_4']))
rmse_Model_5 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_5']))
rmse_Model_6 = np.sqrt(mean_squared_error(valid['Value'], valid['Predict_Model_6']))

# Compute MAE for each combined forecast of Model A and Model B with AR(p_best) and the other ARIMA and SARIMA models
mae_combined_Model_A = mean_absolute_error(valid['Value'], valid['Combined_Predict_Model_A'])
mae_combined_Model_B = mean_absolute_error(valid['Value'], valid['Combined_Predict_Model_B'])
mae_Model_A = mean_absolute_error(valid['Value'], valid['Predict_Model_A'])
mae_Model_B = mean_absolute_error(valid['Value'], valid['Predict_Model_B'])
mae_Model_1 = mean_absolute_error(valid['Value'], valid['Predict_Model_1'])
mae_Model_2 = mean_absolute_error(valid['Value'], valid['Predict_Model_2'])
mae_Model_3 = mean_absolute_error(valid['Value'], valid['Predict_Model_3'])
mae_Model_4 = mean_absolute_error(valid['Value'], valid['Predict_Model_4'])
mae_Model_5 = mean_absolute_error(valid['Value'], valid['Predict_Model_5'])
mae_Model_6 = mean_absolute_error(valid['Value'], valid['Predict_Model_6'])


# Compute MAPE for each combined forecast of Model A and Model B with AR(p_best) and the other ARIMA and SARIMA models
mape_combined_Model_A = compute_mape(valid['Value'], valid['Combined_Predict_Model_A'])
mape_combined_Model_B = compute_mape(valid['Value'], valid['Combined_Predict_Model_B'])
mape_Model_A = compute_mape(valid['Value'], valid['Predict_Model_A'])
mape_Model_B = compute_mape(valid['Value'], valid['Predict_Model_B'])
mape_Model_1 = compute_mape(valid['Value'], valid['Predict_Model_1'])
mape_Model_2 = compute_mape(valid['Value'], valid['Predict_Model_2'])
mape_Model_3 = compute_mape(valid['Value'], valid['Predict_Model_3'])
mape_Model_4 = compute_mape(valid['Value'], valid['Predict_Model_4'])
mape_Model_5 = compute_mape(valid['Value'], valid['Predict_Model_5'])
mape_Model_6 = compute_mape(valid['Value'], valid['Predict_Model_6'])


# Print all the results
print(f"Combined Linear Trend + Additive Seasonality Model A with AR(2) on residuals - RMSE: {rmse_combined_Model_A:.2f}, MAE: {mae_combined_Model_A:.2f}, MAPE: {mape_combined_Model_A:.2f}%")
print(f"Combined Exponential Trend + Multiplicative Seasonality Model B with AR(2) on residuals - RMSE: {rmse_combined_Model_B:.2f}, MAE: {mae_combined_Model_B:.2f}, MAPE: {mape_combined_Model_B:.2f}%")
print(f"Combined Linear Trend + Additive Seasonality Model A with no residual treatment - RMSE: {rmse_Model_A:.2f}, MAE: {mae_Model_A:.2f}, MAPE: {mape_Model_A:.2f}%")
print(f"Combined Exponential Trend + Multiplicative Seasonality Model B with no residual treatment - RMSE: {rmse_Model_B:.2f}, MAE: {mae_Model_B:.2f}, MAPE: {mape_Model_B:.2f}%")
print(f"ARIMA Model 1 - RMSE: {rmse_Model_1:.2f}, MAE: {mae_Model_1:.2f}, MAPE: {mape_Model_1:.2f}%")
print(f"ARIMA Model 2 - RMSE: {rmse_Model_2:.2f}, MAE: {mae_Model_2:.2f}, MAPE: {mape_Model_2:.2f}%")
print(f"SARIMA Model 3 - RMSE: {rmse_Model_3:.2f}, MAE: {mae_Model_3:.2f}, MAPE: {mape_Model_3:.2f}%")
print(f"SARIMA Model 4 - RMSE: {rmse_Model_4:.2f}, MAE: {mae_Model_4:.2f}, MAPE: {mape_Model_4:.2f}%")
print(f"SARIMA Model 5 - RMSE: {rmse_Model_5:.2f}, MAE: {mae_Model_5:.2f}, MAPE: {mape_Model_5:.2f}%")
print(f"SARIMA Model 6 - RMSE: {rmse_Model_6:.2f}, MAE: {mae_Model_6:.2f}, MAPE: {mape_Model_6:.2f}%")
```
